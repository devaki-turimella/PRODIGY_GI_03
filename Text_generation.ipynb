```python

import random


def build_markov_chain(text):

    words = text.split()
    markov_chain = {}

    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]

        # If we've seen this word before, add to its list of "next words"
        if current_word in markov_chain:
            markov_chain[current_word].append(next_word)
        else:
            markov_chain[current_word] = [next_word]

    return markov_chain


def generate_text(markov_chain, start_word, num_words=30):
    """
    Generates text by starting at 'start_word' and repeatedly choosing
    a random next word based on the markov_chain dictionary.
    """
    # If the start word was never seen in training, pick a random known word
    if start_word not in markov_chain:
        start_word = random.choice(list(markov_chain.keys()))

    current_word = start_word
    generated_words = [current_word]

    for _ in range(num_words - 1):
        # Get possible next words for the current word
        next_words = markov_chain.get(current_word)

        # If current word has no known followers, stop early
        if not next_words:
            break

        # Randomly choose the next word from the possibilities
        next_word = random.choice(next_words)
        generated_words.append(next_word)
        current_word = next_word

    return " ".join(generated_words)


def main():
    print("=== Markov Chain Text Generator ===\n")

    # Sample training text (you can replace this with any text you like)
    sample_text = """
    the sun was shining brightly over the calm blue ocean.
    the sun rose slowly over the quiet mountains.
    the ocean waves crashed gently on the golden sand.
    the mountains stood tall against the clear blue sky.
    the sky turned orange as the sun began to set.
    """

    # Let the user provide their own text, or use the sample text
    user_choice = input(
        "Do you want to enter your own training text? (yes/no): "
    ).strip().lower()

    if user_choice == "yes":
        print("\nEnter your training text (a paragraph works best):")
        training_text = input("> ")
    else:
        training_text = sample_text
        print("\nUsing sample training text.")

    # Build the Markov chain from the training text
    chain = build_markov_chain(training_text)

    # Ask the user how many words to generate
    try:
        num_words = int(input("\nHow many words should be generated? (e.g. 20): "))
    except ValueError:
        num_words = 20
        print("Invalid input, defaulting to 20 words.")

    # Ask for a starting word
    start_word = input("Enter a starting word (or press Enter for random): ").strip()
    if start_word == "":
        start_word = random.choice(list(chain.keys()))

    # Generate and display the text
    result = generate_text(chain, start_word, num_words)

    print("\n--- Generated Text ---")
    print(result)


if __name__ == "__main__":
    main()
```
